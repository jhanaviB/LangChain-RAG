```python
from langchain_community.document_loaders import Docx2txtLoader
from langchain_text_splitters.markdown import MarkdownHeaderTextSplitter
from langchain_text_splitters.character import CharacterTextSplitter
from langchain_openai.embeddings import OpenAIEmbeddings

import numpy
```

```python
loader_docx = Docx2txtLoader("Introduction_to_Data_and_Data_Science-2.docx")
pages = loader_docx.load()
```

```python
pages
```

```python
md_splitter = MarkdownHeaderTextSplitter(headers_to_split_on = [("#", "Course Title"),
                                                                ("##","Lecture Title")])
```

```python
pages_md_split = md_splitter.split_text(pages[0].page_content)
```

```python
for i in range(len(pages_md_split)):
    pages_md_split[i].page_content = ' '.join(pages_md_split[i].page_content.split())

for page in pages_md_split:
    page.metadata['author'] = "jhanavi"
    
char_splitter = CharacterTextSplitter(
    separator = ".",
    chunk_size = 500,
    chunk_overlap = 50
)

pages_char_split = char_splitter.split_documents(pages_md_split)
```

```python
pages_char_split
```

```python
embedding = OpenAIEmbeddings(model = "text-embedding-ada-002")
```

```python
vector1 = embedding.embed_query(pages_char_split[0].page_content)
vector2 = embedding.embed_query(pages_char_split[1].page_content)

vector18 = embedding.embed_query(pages_char_split[15].page_content)
```

```python
vector1
```

```python
numpy.dot(vector1,vector2),numpy.dot(vector1,vector18),numpy.dot(vector2,vector18)
```
