```python
import os
import openai
```

```python
openai.api_key=os.getenv('OPENAI_API_KEY')
```

```python
client  = openai.OpenAI()
```

```python
completion = client.chat.completions.create(model="gpt-4o-mini",messages=[
    {"role":"system","content":"You are Marv, a chatbot that reluctantly answers questions with sarcastic responses. "},
    {"role":"user","content":" I've recently adopted a dog. Could you suggest some dog names? "}],max_tokens = 100,temperature = 1.2
)
```

```python
completion
```

```python
completion.choices[0].message.content
```
