```python
from langchain_community.document_loaders import Docx2txtLoader
from langchain_text_splitters.markdown import MarkdownHeaderTextSplitter
from langchain_text_splitters.character import CharacterTextSplitter
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain_core.documents import Document

import numpy
```

```python
loader_docx = Docx2txtLoader("Introduction_to_Data_and_Data_Science-2.docx")
pages = loader_docx.load()
```

```python
pages
```

```python
md_splitter = MarkdownHeaderTextSplitter(headers_to_split_on = [("#", "Course Title"),
                                                                ("##","Lecture Title")])
```

```python
pages_md_split = md_splitter.split_text(pages[0].page_content)
```

```python
for i in range(len(pages_md_split)):
    pages_md_split[i].page_content = ' '.join(pages_md_split[i].page_content.split())

for page in pages_md_split:
    page.metadata['author'] = "jhanavi"
    
char_splitter = CharacterTextSplitter(
    separator = ".",
    chunk_size = 500,
    chunk_overlap = 50
)

pages_char_split = char_splitter.split_documents(pages_md_split)
```

```python
pages_char_split
```

```python
embedding = OpenAIEmbeddings(model = "text-embedding-ada-002")
```

```python
vectorstore = Chroma.from_documents(documents = pages_char_split,
                                    embedding = embedding, 
                                    persist_directory = "./intro-to-ds-lectures")
```

```python
vectorstore_from_directory = Chroma(persist_directory = "./intro-to-ds-lectures",
                                    embedding_function = embedding)
```

```python
vectorstore_from_directory.get("f52c63f2-ab89-4355-b6c6-f3408a72996b")
```

```python
added_document = Document(page_content = "More importantly, it will be sufficient for your need to create quick and accurate analyses. However, if your theoretical preparation is strong enough, you will find yourself restricted by software. Knowing a programming language such as R and Python, gives you the freedom to create specific, ad-hoc tools for each project you are working on",
                          metadata={'author': 'jhanavi'})
```

```python
vectorstore_from_directory.add_documents([added_document])
```
