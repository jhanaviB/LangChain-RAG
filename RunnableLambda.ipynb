```python
from langchain_core.runnables import RunnableLambda
```

```python
runnable_sum = RunnableLambda(lambda x:sum(x))
```

```python
runnable_sum.invoke([1,3,5])
```

```python
runnable_square = RunnableLambda(lambda x:x*x)
```

```python
runnable_square.invoke(9)
```

```python
runnable_combined = runnable_sum | runnable_square
```

```python
runnable_combined.invoke([1,3,5])
```
