```python
%load_ext dotenv
%dotenv
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import CommaSeparatedListOutputParser
```

```python
chat_template = ChatPromptTemplate.from_messages([('human',"I've recently adopted a {pet} which is a {breed}. Could you suggest some training tips?")])
```

```python
chat = ChatOpenAI(model_name = 'gpt-4',
                  seed = 365,
                  temperature = 0,
                  max_tokens = 100)
```

```python
list_output_parser = CommaSeparatedListOutputParser()
```

```python
chain = chat_template|chat|list_output_parser
```

```python
%%time
chain.invoke({'pet':'dog','breed':'golden shepherd'})
```

```python
%%time
chain.invoke({'pet':'cat','breed':'shitzu'})
```

```python
%%time
chain.batch([{'pet':'dog','breed':'golden shepherd'},{'pet':'cat','breed':'shitzu'}])
```

```python
response = chain.stream({'pet':'dog','breed':'golden shepherd'})
```
