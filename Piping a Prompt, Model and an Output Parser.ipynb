```python
%load_ext dotenv
%dotenv
```

```python
from langchain_openai import ChatOpenAI

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import CommaSeparatedListOutputParser
```

```python
list_instructions = CommaSeparatedListOutputParser().get_format_instructions()
```

```python
list_instructions
```

```python
chat_template = ChatPromptTemplate.from_messages([('human',"I've recently adopted a {pet}. Could you suggest three {pet} names? \n" + list_instructions)])
```

```python
print(chat_template.messages[0].prompt.template)
```

```python
list_output_parser = CommaSeparatedListOutputParser()
```

```python
chat_template_result = chat_template.invoke({'pet':'dog'})
```

```python
chat = ChatOpenAI(model_name = 'gpt-4',
                  seed = 365,
                  temperature = 0,
                  max_tokens = 100)
```

```python
chat_result = chat.invoke(chat_template_result).content
```

```python
list_output_parser.invoke(chat_result)
```

```python
chain = chat_template|chat|list_output_parser
```

```python
chain.invoke({'pet':'dog'})
```
